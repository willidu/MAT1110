```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 5.4.1
N = 100
x, y = np.zeros((2, N+1))
n = np.linspace(0, N, N+1, dtype=int)

for i in range(N):
    x[i+1] = 0.6*x[i] - 0.6*y[i] + 0.2
    y[i+1] = 0.6*x[i] + 0.6*y[i] + 1

print(f'Fixpoint: (x, y) = ({x[-1]:.2f}, {y[-1]:.2f})')

plt.plot(n, x, label='x')
plt.plot(n, y, label='y')
plt.legend()
plt.show()
```

```python
# 5.4.2
N = 200
x, y = np.zeros((2, N+1))
n = np.linspace(0, N, N+1, dtype=int)

x[0] = 20
y[0] = 2000

for i in range(N):
    x[i+1] = 0.9*x[i] + 0.01*y[i] - 10
    y[i+1] = -1.01*x[i] + y[i] + 300

print(f'Equilibrium point: (x, y) = ({x[-1]:.0f}, {y[-1]:.0f})')

plt.plot(n, x, label='x')
plt.plot(n, y, label='y')
plt.legend()
plt.show()
```

```python
# 5.4.7
N = 30
t = np.linspace(0, N, N+1)
x = np.zeros(N+1)
x[0] = np.sqrt(2)

def f(x):
    return x**2 + x - 2

for i in range(N):
    x[i+1] = f(x[i])

plt.plot(t, x)
plt.show()
```
