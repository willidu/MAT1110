```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = 3

t = np.linspace(-b, b, 1001)
r = np.asarray([a*np.arcsinh(t/a), np.sqrt(t**2+a**2)])

plt.plot(r[0], r[1], label='r(t)')
plt.legend()
plt.show()
```

```python
for a, b, c in ((1, 1, 1), (-1, 3, -1)):
    d = -a

    t = np.linspace(-5, 5, 21)
    x, y = np.meshgrid(t, t)

    u, v = a*x + b*y, c*x + d*y
    phi = c*x**2 - 2*a*x*y - b*y**2

    plt.quiver(x, y, u, v)
    plt.contour(x, y, phi)
    plt.show()
```
